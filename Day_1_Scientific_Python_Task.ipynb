```python
df.drop_duplicates(inplace = True)
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.info()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df["Year"] = df["Order_Date"].dt.year
df["Year"]
```

```python
df.drop("Year", axis = 1, inplace = True)
```

```python
Monthly_Sales_Profit = df.groupby("Month")["Sales_Profit"].sum()
```

```python
Monthly_Sales_Profit.plot(kind = "line")
```

```python
Monthly_Sales_Profit.plot(kind = "bar")
```
