```python
df.drop_duplicates(inplace = True)
```

```python
df.rename(columns = {"STATE": "STATE_NAME"}, inplace = True)
```

```python
df.rename(columns = {"CITY" : "REGION"}, inplace = True)
```

```python
df["ORDERDATE"]
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])
```

```python
df["ORDERDATE"]
```

```python
df["YEAR"] = df["ORDERDATE"].dt.year
df["YEAR"]
```

```python
df["MONTH"] = df["ORDERDATE"].dt.month
df["MONTH"]
```

```python
df["DAY"] = df["ORDERDATE"].dt.day
df["DAY"]
```

```python
df["PROFIT_EST"] = df["SALES"] * 0.2
```

```python
df["PROFIT_EST"]
```

```python
df.drop("PROFIT_EST", axis = 1, inplace = True)
```

```python
df["FULLNAME"] = df["CONTACTFIRSTNAME"] + " " +df["CONTACTLASTNAME"]
```

```python
df["FULLNAME"]
```

```python
monthly_sales = df.groupby("MONTH_ID")["SALES"].sum()
```

```python
monthly_sales.plot(kind = "line")
```

```python
monthly_sales.plot(kind = "bar")
```

```python
df.groupby("COUNTRY")["SALES"].sum()
```

```python
order_per_year = df.groupby("YEAR_ID")["ORDERNUMBER"].nunique()
print(order_per_year)
```

```python
avg_sales_product = df.groupby("PRODUCTCODE")["SALES"].mean()
print(avg_sales_product)
```

```python
df["DEALSIZE"] = df["DEALSIZE"].str.upper()
df["DEALSIZE"]
```
