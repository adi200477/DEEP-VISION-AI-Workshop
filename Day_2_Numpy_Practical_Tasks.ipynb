```python
# 11
arr1 = np.arange(1,25)
print("Orginal Array")
print(arr1,"\n")
print("Reshaped (6, 4)")
print(arr1.reshape(6,4),"\n")
print("Reshaped (2, 3, 4)")
print(arr1.reshape(2, 3, 4),"\n")
```

```python
# 12
arr2 = np.random.randint(25, 100, 20)
print(arr2, "\n")
print("Mean")
print(np.mean(arr2), "\n")
print("Median")
print(np.median(arr2), "\n")
print("Standard Deviation")
print(np.std(arr2), "\n")
```

```python
# 13
print(arr2[arr2>50])
```
