## Arthmetic Operations

```python
arr1 = np.array([4, 6, 2])
arr2 = np.array([2, 8, 7])

sum_arr = arr1 + arr2
sub_arr = arr1 - arr2
prod_arr = arr1 * arr2
div_arr = arr1 / arr2
exp_arr = np.exp(arr1)
log_arr = np.log(arr2)

print(sum_arr)
print(sub_arr)
print(prod_arr)
print(div_arr)
print(exp_arr)
print(log_arr)
```

```python
z = np.array([[1, 2, 3], [0, 8, 9], [10, 20,23]])

print(z)
print(z[:,0])
print(z[0, :])
print(z[1, 0])
print(z[2, 1])
print(z[2, :])
print(z[:, 1])
print(z[2, 2])
print(z[1, 2])
```

## Image Preprocessing

```python
from PIL import Image
```

```python
image = Image.open("/content/luffy.jpg")
image
```

```python
image_resized = image.resize((400, 200))
image_resized
```

## Normalization

```python
import numpy as np
image_arr = np.array(image)
image_arr
```

```python
image_arr = np.array(image) / 255.0   # Normalization
image_arr
```

## Color Conversion

```python
from cv2 import cvtColor, COLOR_BGR2GRAY
image_np = np.array(image)
grayscale_image = cvtColor(image_np, COLOR_BGR2GRAY)
grayscale_image
```

## Data Augmentation

```python
from PIL import Image, ImageEnhance, ImageOps

flipped = ImageOps.mirror(image)
flipped
```

```python
rotated = image.rotate(275)
rotated
```

```python
enhancer = ImageEnhance.Brightness(image)
brighter = enhancer.enhance(1.5)
brighter
```

```python
contrast = ImageEnhance.Contrast(image).enhance(2)
contrast
```
